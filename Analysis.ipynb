## What products are sold together often?

```python
results = all_data.loc[:,["Order ID","Product","Quantity Ordered"]].sort_values(by="Order ID")
results.head()
```

```python
product_list=[product for product in results['Product']]
product_list = [ product for product in set(product_list)]
print(product_list)
```

```python
product_dict={}
for k in results.index:
    product = results['Product'][k]
    id = results['Order ID'][k]
    product_dict.setdefault(id, []).append(results['Product'][k])
```

```python
purchase_dict={}
for product in product_list:
    for id in product_dict:
        if (product in product_dict[id]) and (len(set(product_dict[id])) > 1):
            purchase_dict.setdefault(product, []).extend(product_dict[id])
```

```python
for item in purchase_dict:
    for sub_item in purchase_dict[item]:
        if sub_item==item:
            purchase_dict[item].pop(purchase_dict[item].index(sub_item))
```

```python
for item in purchase_dict:
    temp = purchase_dict[item]
    purchase_dict[item]={}
    for prod in temp:
        purchase_dict[item][prod] = 0
        while prod in temp:
            purchase_dict[item][prod]+=1
            temp.pop(temp.index(prod))
```

```python
purch_dict = {}
for item in purchase_dict:
    print(item)
    print(purchase_dict[item])
    for sub_item in purchase_dict[item]:
        purch_dict[item,sub_item]=purchase_dict[item][sub_item]
```

```python
purchase_history ={'item': purch_dict.keys(), 'number of times': purch_dict.values()}
data_history = pd.DataFrame.from_dict(purchase_history)
```

```python
data_history = data_history.sort_values(by='number of times',ascending=False)
```

```python
temp = data_history
data_history['purchased'] = data_history['item'].apply(lambda x: x[1])
data_history['item'] = data_history['item'].apply(lambda x: x[0])
#
data_history
```

```python
data_history = data_history.loc[:,["item","purchased","number of times"]].sort_values(by="number of times",ascending=False)
```

```python
temp = data_history
for i in temp.index:
    for j in temp.index:
        try:
            if (temp['item'][i] is temp['purchased'][j]) and (temp['purchased'][i] is temp['item'][j]):
                temp = temp.drop(j)
        except KeyError:
            pass
for i in temp.index:
    if temp['item'][i] == temp['purchased'][i]:
        temp = temp.drop(i)
temp = temp.sort_values(by="number of times",ascending=False)
```

```python
temp = temp.drop('index',axis=1)
```

```python
data_history = temp.groupby('item').max()
data_history.index = data_history.index.map(str)
```
