```python
all_data = pd.read_csv("all_months_data.csv")
all_data.head()
```

# adding helpful columns

clean NaN 

```python
for i in all_data.index:
    if pd.isnull(all_data['Quantity Ordered'][i]) or all_data['Quantity Ordered'][i] =='Quantity Ordered' :
        all_data = all_data.drop(i) 
all_data.to_csv("all_months_data.csv")
```

```python
all_data = all_data.drop('Unnamed: 0',axis=1)
all_data.head()
```

add a month column

```python
all_data['month']=all_data['Order Date'].str[0:2]
all_data['month'] = all_data['month'].astype('int32')
all_data
```

```python
all_data= all_data.sort_values(by="month")
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int')
all_data['Price Each'] = all_data['Price Each'].astype('float')
```

adding a profit column

```python
all_data['Profit']=all_data['Quantity Ordered'] * all_data['Price Each']
all_data
```

```python
# Whats was the best month for sales? April
```

```python
print(all_data.loc[:,["month","Profit"]])
```

```python
monthly_profits = {}
for k in all_data.index:
    if all_data['month'][k] not in monthly_profits: 
        monthly_profits[all_data['month'][k]]=all_data['Profit'][k]
    else:monthly_profits[all_data['month'][k]]+=all_data['Profit'][k]
```
